```python
from shapely import geometry
import geopandas as gpd
import pandas as pd
import os
import numpy as np
```

```python
coord_list = pd.read_csv("Coordinates-Binning3.csv")
def Coord_lister(Loc_Column):
    crs = {'init': 'epsg:4326'}
    x = Loc_Column.to_list()
    pointList = [item for item in x if str(item) != 'nan']
    pointL = []
    for x in pointList:
        y = eval(x)
        pointL.append(y)
    poly = geometry.Polygon(pointL)
    spoly = gpd.GeoSeries([poly],crs=crs)
    return spoly
He_Shop = Coord_lister(coord_list["HE shop"])
Crusher = Coord_lister(coord_list["Crusher"])
Annex_A = Coord_lister(coord_list["Annex A"])
Carmen_pit = Coord_lister(coord_list["Carmen Pit"])

Location_list = [He_Shop,Crusher, Annex_A, Carmen_pit]
Location_list_names = ["He_Shop","Crusher", "Annex_A", "Carmen Pit"]
```

```python
file = '33854 Feb.csv'
crs = {'init': 'epsg:4326'}

#Create a geoseries holding the single polygon. Coordinates in counter-clockwise order
#pointList = [(10.325852, 123.726840), (10.331532, 123.718828), (10.337319, 123.724241), (10.341579, 123.727525), (10.340266, 123.732830), (10.333982, 123.733516)]

#Create geodataframe of points
dfcsv = pd.read_csv(file)
dfcsv
```

```python
geometry = [geometry.Point(xy) for xy in zip(dfcsv.Latitude, dfcsv.Longitude)]
dfpoints = gpd.GeoDataFrame(dfcsv, crs=crs, geometry=geometry)
dfpoints
```

```python
#Create a subset dataframe of points within the polygon
for x,y in zip(Location_list, Location_list_names):
    dfpoints.loc[dfpoints.within(x.geometry.iloc[0]), "Location"] = y
dfpoints.pivot_table(index = "Location")
dfpoints.loc[dfpoints["Location"].isna(), "Location"] = "Others"
dfpoints
```

```python
print("Data Successfully Combined")
export_state = False
while not export_state:
    export_path = input("Please input export path directory \n")
    if os.path.exists(export_path) == False:
        print("Please input a correct file path")
    else:
        export_state = True
csv_fname = input("Please input filename")
export_fpath = export_path + "\\" + csv_fname + ".csv"
print(export_fpath)
dfpoints.to_csv(path_or_buf = export_fpath, index = False)
```
